```python
import pandas as pd

df = pd.read_csv(r"E:\KarmixTech_Executive_BI_Project\Executive_BI_Sales_Dataset.csv")
df.info()
df.describe()
df.isnull().sum()
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])
df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month_name()
```

```python
# Total Revenue & Profit
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
profit_margin = (total_profit / total_sales) * 100

# Profit by Category (loss-making category dhundo)
df.groupby('Category')['Profit'].sum().sort_values()

# Year-over-Year growth
df.groupby('Year')['Sales'].sum()

# Region-wise performance
df.groupby('Region')[['Sales','Profit']].sum()
```

```python
print("Total Sales:", df['Sales'].sum())
print("Total Profit:", df['Profit'].sum())
print("Profit Margin %:", (df['Profit'].sum()/df['Sales'].sum())*100)

print("\nProfit by Category:")
print(df.groupby('Category')['Profit'].sum().sort_values())

print("\nSales by Year:")
print(df.groupby('Year')['Sales'].sum())

print("\nRegion-wise:")
print(df.groupby('Region')[['Sales','Profit']].sum())
```
